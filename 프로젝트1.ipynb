### 3-2 데이터 분석하기

    필수 문제1 : 월별 매출액 추세(sales_data를 이용합니다.)
    월별 매출액 추세를 집계하고 그 결과를 시각화해봅니다.
    힌트: groupby() 함수를 사용하세요.

```python
# 활용할 데이터(sales_data)를 temp_var로 정의하기
# [[YOUR CODE]]
temp_var = sales_data
```

```python
# temp_var의 date열을 만들고, sales_date의 결제일시 중 앞에서 7번째 값까지 할당
# [[YOUR CODE]]
temp_var['date'] = sales_data['결제일시'].str[:7]
temp_var
```

```python
# salse_data의 결제금액에 포함된 쉼표(,) 지우기
# [[YOUR CODE]]
sales_data['결제금액'] = sales_data['결제금액'].str.replace(',', '',)
temp_var
```

```python
# salse_data의 결제금액을 숫자형태로 바꾸기
# [[YOUR CODE]]
sales_data['결제금액'].astype(int)
```

```python
# temp_var의 payment열을 date 기준으로 집계하기
# [[YOUR CODE]]
temp_var1 = temp_var['결제금액'].groupby(temp_var['date']).count()
temp_var1
```

```python
# 집계 결과를 시각화하기
# [[YOUR CODE]]
plt.figure(figsize=(10,5))
plt.bar(temp_var1.index, temp_var1)
plt.title('월별 결제금액')
plt.xticks(rotation=45)
plt.ylabel('판매건수')
plt.show()
```

    필수 문제2: 가장 인기 있는 음식/음료(salses_data를 이용합니다.)
    여러 메뉴들(음식,음료)을 판매 개수 순으로 정렬하고 상위 10개를 뽑아 결과를 시각화해봅니다.
    힌트: value_counts() 함수를 사용하세요.

```python
# temp_var의 상품명을 value_counts를 이용해 집계하기
# 주의! value_counts로 바로 집계하면 판매취소와 결제취소가 함께 집계됩니다.
# 따라서 '판매수량' 칼럼 값이 '-1'인 데이터를 제외하고 집계해야 합니다.
# 힌트 코드 : temp_var[temp_var['판매수량']==-1]
# [[YOUR CODE]]

sold_goods = temp_var[temp_var['판매수량']!= -1].value_counts()

sold_goods
```

```python
# 집계 결과를 데이터프레임으로 변환하고, 상위 10개 선정하기
# [[YOUR CODE]]
sold_goods = pd.DataFrame(sold_goods)

top_goods= sold_goods[:10]
top_goods
```

```python
# 한글이 포함된 시각화를 위해 Matplotlib 설정
# [[YOUR CODE]]

import koreanize_matplotlib
```

```python
# 집계 결과를 시각화하기
# [[YOUR CODE]]
top_goods[:10].plot(kind='bar')
plt.ylabel('개')
plt.title('판매 개수')
plt.show()
```

```python
# temp_var의 payment열을 상품명을 기준으로 집계하기
# [[YOUR CODE]]
item_payment = temp_var['결제금액'].groupby(temp_var['상품명']).count()
item_payment
```

```python
# item_payment의 값을 내림차순으로 정렬하고 상위 10개 도출하기
# [[YOUR CODE]]
top10_item= item_payment.sort_values(ascending=False).head(10)
top10_item
```

```python
# 집계 결과 시각화하기
# [[YOUR CODE]]
plt.figure(figsize=(10,5))
plt.bar(top10_item.index, top10_item)
plt.title('top10 품목')
plt.xticks(rotation=45)
plt.ylabel('판매건수')
plt.show()
```

    필수 문제3: 가장 많이 팔린 입장권 종류(entry_data를 이용합니다.)
    여러 입장권 종류를 판매 매출(판매 개수 x 금액) 순으로 정렬하고 상위 10개를 뽑아 결과를 시각화해봅니다.

```python
# 활용할 데이터(entry_data)를 temp_var로 정의하기
# [[YOUR CODE]]
temp_var62 = entry_data
```

```python
# temp_var의 금액을 숫자형태로 변환하기
# [[YOUR CODE]]
temp_var62['금액'].astype(int)
```

```python
# temp_var의 금액을 요금제명을 기준으로 집계하기
# [[YOUR CODE]]
tickets =temp_var62['금액'].groupby(temp_var62['요금제명']).count()
tickets
```

```python
# tickets의 값을 내림차순으로 정렬하고 상위 10개 도출하기
# [[YOUR CODE]]
Top_tickets=tickets.sort_values(ascending=False).head(10)
Top_tickets
```

## item_data의 커피 분류중 제일 많이 시킨 커피는?

```python
# 내 Key Question을 중심으로 다양한 분석을 진행해봅시다.
# 아이템 정보 확인
# [[YOUR CODE]]
item_data.info()
```

```python
# [[YOUR CODE]]
# item 에 대한것 정의
var = item_data
var.head()
```

```python
# [[YOUR CODE]]
# 문자열을 정수로 바꾸기 위해 쉼표와 원 제거
var['판매금액']=var['판매금액'].str.replace(',','')
var['판매금액']=var['판매금액'].str.replace('원','')
var
```

```python
# [[YOUR CODE]]
# var type 정수형으로 변경
var['판매금액'].astype(int)
```

```python
# [[YOUR CODE]]
#상품명 확인
var['분류'].value_counts()
```

```python
# [[YOUR CODE]]
#상품명을 기준으로 분류가 커피인것만 집계
coffee = var[var['분류']=='커피'].groupby(var['상품명']).count()
coffee
```

```python
# [[YOUR CODE]]
#커피 칼럼확인하고,  칼럼 삭제
# coffee.columns
#구분칼럼을 판매건수로 칼럼명변경
coffee1 = coffee.drop([coffee.columns[1],coffee.columns[2],coffee.columns[3]], axis=1)

coffee1.rename(columns={'구분':'판매건수'}, inplace = True)

coffee1
```

```python
# [[YOUR CODE]]
coffee1.plot(kind='bar')
plt.ylabel('개')
plt.title('판매 개수')
plt.xticks(rotation=90)
plt.show()
```
